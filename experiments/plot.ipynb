```python
import matplotlib.pyplot as plt
import pandas as pd

exp_1=pd.read_csv("Experiment_LN.csv")  # Metric for Instance Normalization mentioned in paper
exp_2=pd.read_csv("Experiment_IN.csv") # Metrics for Instance Normalization using tensorflow
exp_3=pd.read_csv("Experiment_NON.csv") # Metrics for No Normalization on last layer
exp_4=pd.read_csv("Experiment_1.csv") # Metrics for Batch Normalization
```

```python
exp_1,exp_2,exp_3,exp_4
```

```python
fig, ax = plt.subplots()
ax.plot(exp_1["epoch"], exp_1["loss"], color='#FF6B6B', linestyle='--', marker='o', markersize=3, label='LN')
ax.plot(exp_2["epoch"], exp_2["loss"], color='#4ECDC4', linestyle='-', marker='s', markersize=3, label='IN')
ax.plot(exp_3["epoch"], exp_3["loss"], color='#FFB86C', linestyle='-', marker='^', markersize=3, label='NoN')
ax.plot(exp_4["epoch"], exp_4["loss"], color='#BD93F9', linestyle='-', marker='D', markersize=3, label='BN')


legend = ax.legend(loc='upper right', shadow=True, fontsize='x-small')
ax.set_ylim([0.2,0.7])
ax.set_xlabel('Epoch')
ax.set_ylabel('Total Loss')
ax.set_title('Traaining Loss')
fig.savefig("Train_loss.png")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(exp_1["epoch"], exp_1["val_loss"], color='#FF6B6B', linestyle='--', marker='o', markersize=3, label='LN')
ax.plot(exp_2["epoch"], exp_2["val_loss"], color='#4ECDC4', linestyle='-', marker='s', markersize=3, label='IN')
ax.plot(exp_3["epoch"], exp_3["val_loss"], color='#FFB86C', linestyle='-', marker='^', markersize=3, label='NoN')
ax.plot(exp_4["epoch"], exp_4["val_loss"], color='#BD93F9', linestyle='-', marker='D', markersize=3, label='BN')


legend = ax.legend(loc='upper right', shadow=True, fontsize='x-small')
ax.set_ylim([0.2,0.7])
ax.set_xlabel('Epoch')
ax.set_ylabel('Total Loss')
ax.set_title('Validation Loss')
fig.savefig("Val_loss.png")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(exp_1["epoch"], exp_1["val_dendrites_iou_score"], color='#FF6B6B', linestyle='--', marker='o', markersize=3, label='LN')
ax.plot(exp_2["epoch"], exp_2["val_dendrites_iou_score"], color='#4ECDC4', linestyle='-', marker='s', markersize=3, label='IN')
ax.plot(exp_3["epoch"], exp_3["val_dendrites_iou_score"], color='#FFB86C', linestyle='-', marker='^', markersize=3, label='NoN')
ax.plot(exp_4["epoch"], exp_4["val_dendrites_iou_score"], color='#BD93F9', linestyle='-', marker='D', markersize=3, label='BN')

legend = ax.legend(loc='lower right', shadow=True, fontsize='x-small')
ax.set_ylim([0.2,0.7])
ax.set_xlabel('Epoch')
ax.set_ylabel('IoU Score')
ax.set_title('Validation IoU(dendrites)')
fig.savefig("IoU_den_Val.png")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(exp_1["epoch"], exp_1["val_spines_iou_score"], color='#FF6B6B', linestyle='--', marker='o', markersize=3, label='LN')
ax.plot(exp_2["epoch"], exp_2["val_spines_iou_score"], color='#4ECDC4', linestyle='-', marker='s', markersize=3, label='IN')
ax.plot(exp_3["epoch"], exp_3["val_spines_iou_score"], color='#FFB86C', linestyle='-', marker='^', markersize=3, label='NoN')
ax.plot(exp_4["epoch"], exp_4["val_spines_iou_score"], color='#BD93F9', linestyle='-', marker='D', markersize=3, label='BN')

legend = ax.legend(loc='upper right', shadow=True, fontsize='x-small')
ax.set_ylim([0.2,0.7])
ax.set_xlabel('Epochs')
ax.set_ylabel('IoU Score')
ax.set_title('Validation IoU(Spines)')
fig.savefig("IoU_spine_Val.png")

plt.show()
```
